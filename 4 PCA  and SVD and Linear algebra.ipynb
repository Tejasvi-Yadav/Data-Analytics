# principal component analysis

```python
data=np.array([[67,   3,   1089,   56],[48,   1,   1234,   51],[87,7,1256,59],[76,5,1052,58],[71,9,1114,62],[71,6,1068,50],[73,8,1139,49],[70,4,1139,60]])
```

```python
data
```

```python
data.shape
```

```python
#import class
from sklearn.decomposition import PCA

#create the object
pca = PCA()

x_scaled = scaler.fit_transform(data)
#apply the algorithm
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
plt.xlabel('Feature Number')
plt.ylabel('variance')
plt.grid()
plt.bar(range(4) ,pca.explained_variance_ratio_)
plt.plot(range(4) ,pca.explained_variance_ratio_, marker='.' , color='red')

```

```python
pca=PCA(n_components=2) # n_component = choose the smallest number of components that 
#still capture most of the information in your data
```

```python
x_pca=pca.fit_transform(x_scaled)
```
